```python
#!/usr/bin/env python
import subprocess
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
p = subprocess.run("rwfilter ./mysterious-dataset.rw --pass=stdout  --type=all | rwuniq --fields=sip,dip,dport --sort-output --flows --no-columns --column-separator=' ' --no-title",
                    capture_output=True,
                    shell=True
                )
if p.stderr:
    raise p.stderr.decode()

out = p.stdout.decode()
```

```python
usual_servers = []
http_server = []
f = out.splitlines()
for l in f:
    sip,dip,dport,count = l.strip().split(' ')
    if dip == "192.168.0.200":
        http_server.append("sip:{},dip:{},dport:{},count:{}".format(sip,dip,dport,count))
    elif sip == "192.168.0.200":
        continue
    else:
        usual_servers.append("sip:{},dip:{},dport:{},count:{}".format(sip,dip,dport,count))

print(usual_servers)
print(http_server)
```
